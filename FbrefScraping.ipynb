# Extracción de tablas de jugadores

```python
import pandas as pd
import requests

fbref_url = 'https://fbref.com/en/comps/Big5/shooting/players/Big-5-European-Leagues-Stats'
#fbref_url = 'https://fbref.com/en/comps/Big5/2017-2018/stats/players/2017-2018-Big-5-European-Leagues-Stats'
# fbref_url = 'https://fbref.com/en/comps/Big5/2023-2024/stats/players/2023-2024-Big-5-European-Leagues-Stats'
# Replaces comment start and end with blank to find the player table
df = pd.read_html(requests.get(fbref_url).text.replace('<!--', '').replace('-->', ''), attrs={'id':'stats_standard'}, index_col=[0])[0]
#stats_standard
#stats_misc
#stats_possession
#stats_shooting
#stats_passing
#stats_gca
#stats_defense

df.columns = df.columns.get_level_values(1)

df = df.reset_index().rename(columns={'index': 'Player_id'})

df = df[df['Player_id'] != 'Rk']
df
df = df.fillna(0)
df
```

```python
# Guardar el DataFrame en un archivo Excel
nombre_archivo_excel = 'BigFive2425Shooting.xlsx'
df.to_excel(nombre_archivo_excel, index=False)

print(f'DataFrame guardado en {nombre_archivo_excel}')
```

```python
import pandas as pd
import requests
import openpyxl# Add this line to import the requests library
import html5lib

#fbref_url = 'https://fbref.com/en/comps/32/2022-2023/misc/2022-2023-Primeira-Liga-Stats'
#fbref_url = 'https://fbref.com/en/comps/Big5/2017-2018/stats/players/2017-2018-Big-5-European-Leagues-Stats'
# fbref_url = 'https://fbref.com/en/comps/Big5/2017-2018/defense/players/2017-2018-Big-5-European-Leagues-Stats'
#fbref_url = 'https://fbref.com/en/comps/Big5/2022-2023/2022-2023-Big-5-European-Leagues-Stats'
fbref_url = 'https://fbref.com/en/comps/Big5/2023-2024/passing/players/2023-2024-Big-5-European-Leagues-Stats'
# Replaces comment start and end with blank to find the player table
df = pd.read_html(requests.get(fbref_url).text.replace('<!--', '').replace('-->', ''), attrs={'id':'stats_defense'}, index_col=[0])[0]
#stats_standard
#stats_misc
#stats_possession
#stats_shooting
#stats_passing
#stats_gca
#stats_defense

df.columns = df.columns.get_level_values(1)

df = df.reset_index().rename(columns={'index': 'Player_id'})

df = df[df['Player_id'] != 'Rk']

df = df.fillna(0)
df
```
