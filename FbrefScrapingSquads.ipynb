```python
# Hacer scraping para cada una de las 7 estadísticas 
import pandas as pd
import numpy

url_df2324 = "https://fbref.com/en/comps/Big5/2023-2024/shooting/squads/2023-2024-Big-5-European-Leagues-Stats"

df = pd.read_html(url_df2324)[0]
df
```

```python
df = pd.read_html(url_df2324)
df = df[0]  
df
```

```python
df.columns = [' '.join(col).strip() for col in df.columns]

df = df.reset_index(drop=True)
df
```

```python
new_columns = []
for col in df.columns:
    if 'level_0' in col:
        new_col = col.split()[-1]
    else:
        new_col = col
    new_columns.append(new_col)

df.columns = new_columns
df = df.fillna(0)

df.head()
```
