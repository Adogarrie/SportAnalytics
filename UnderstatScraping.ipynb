```python
import requests
from bs4 import BeautifulSoup
import json
import pandas as pd
pd.set_option('display.max_columns', None)
```

```python
%%time

understat_ids = list(range(26985, 27110+1))  # Crea una lista con el rango de 21925 a 23675
# understat_ids.append(25861)
match_data = []

for id in understat_ids:
    base_urls = f'https://understat.com/match/{id}'
#Use requests to get the webpage and BeautifulSoup to parse the page
    res = requests.get(base_urls)
    soup = BeautifulSoup(res.content, 'lxml')
    scripts = soup.find_all('script')
#get only the shotsData
    try:
        strings = scripts[1].string

    # strip unnecessary symbols and get only JSON data
        ind_start = strings.index("('")+2
        ind_end = strings.index("')")
        json_data = strings[ind_start:ind_end]
        json_data = json_data.encode('utf8').decode('unicode_escape')

    #convert string to json format
        data = json.loads(json_data)
    #iterate JSON and extend match_data with a list of dicts
        match_data.extend([d for k in data.keys() for d in data[k]])
        print(f"Obtenido partido {id}")
    except:
        print(f"El id {id} no existe")

global big5
big5 = pd.DataFrame(match_data)

big5['Partido'] = big5['h_team'] + ' ' + big5['h_goals'] + ' ' + '-' + ' ' + big5['a_goals'] + ' ' + big5['a_team']
big5['Competicion'] = 'LaLiga'
```

```python
# Comprobamos que hemos extraido datos de todos los partidos de la temporada 23/24
equipos_laliga = ['Alaves', 'Almeria', 'Athletic Club', 'Atletico Madrid', 'Barcelona', 'Cadiz', 'Celta Vigo', 'Getafe', 
                  'Girona', 'Granada', 'Las Palmas', 'Mallorca', 'Osasuna', 'Real Betis', 'Real Madrid', 'Real Sociedad', 
                  'Rayo Vallecano', 'Sevilla', 'Valencia', 'Villarreal']

epl_teams = [
    'Arsenal', 'Aston Villa', 'Bournemouth', 'Brentford', 'Brighton', 'Burnley', 
    'Chelsea', 'Crystal Palace', 'Everton', 'Fulham', 'Liverpool', 'Luton', 
    'Manchester City', 'Manchester United', 'Newcastle United', 'Nottingham Forest', 
    'Sheffield United', 'Tottenham', 'West Ham', 'Wolverhampton Wanderers'
]

equipos_SerieA = ['AC Milan', 'Roma', 'Atalanta', 'Bologna', 'Cagliari', 'Empoli', 'Fiorentina', 'Frosinone', 
                   'Genoa', 'Verona', 'Inter', 'Juventus', 'Lazio', 'Lecce', 'Monza', 'Napoli', 'Salernitana', 
                   'Sassuolo', 'Torino', 'Udinese']

equipos_bundesliga =  ['Bayern Munich', 'Bayer Leverkusen', 'Borussia Dortmund', 'RasenBallsport Leipzig', 'Freiburg', 
                      'Union Berlin', 'Eintracht Frankfurt', 'Wolfsburg', 'Borussia M.Gladbach', 'Mainz 05',
                      'Hoffenheim', 'VfB Stuttgart', 'Augsburg', 'Werder Bremen', 'Bochum', 'FC Heidenheim', 
                      'Darmstadt', 'FC Cologne']

ligue1_teams = [
    "Paris Saint Germain", "Marseille", "Lens", "Lille", "Monaco", "Lyon", 
    "Rennes", "Nice", "Montpellier", "Clermont Foot", "Strasbourg", "Brest", 
    "Reims", "Nantes", "Toulouse", "Metz", "Le Havre", "Lorient"
]

df_laliga = big5[big5['h_team'].isin(equipos_laliga)]
df_epl = big5[big5['h_team'].isin(epl_teams)]
df_seriea = big5[big5['h_team'].isin(equipos_SerieA)]
df_bundes = big5[big5['h_team'].isin(equipos_bundesliga)]
df_ligue1 = big5[big5['h_team'].isin(ligue1_teams)]
# df_filtrado
total_partidos = df_laliga['Partido'].nunique()+df_epl['Partido'].nunique()+df_seriea['Partido'].nunique()+df_bundes['Partido'].nunique()+df_ligue1['Partido'].nunique()
total_partidos

```

```python
# Añadimos la columna competicion
big5['Competicion'] = None

# Asignar la competición según el equipo
big5.loc[big5['h_team'].isin(epl_teams), 'Competicion'] = 'EPL'
big5.loc[big5['h_team'].isin(equipos_laliga), 'Competicion'] = 'LaLiga'
big5.loc[big5['h_team'].isin(equipos_SerieA), 'Competicion'] = 'SerieA'
big5.loc[big5['h_team'].isin(equipos_bundesliga), 'Competicion'] = 'Bundesliga'
big5.loc[big5['h_team'].isin(ligue1_teams), 'Competicion'] = 'Ligue1'
big5
```

```python
big5['Equipo'] = big5.apply(lambda row: row['h_team'] if row['h_a'] == 'h' else row['a_team'], axis=1)
big5
```

```python
# Renombramos y eliminamos algunas columnas
big5_renamed = big5.rename(columns={'minute': 'Minuto', 'result': 'Suceso','X': 'Posicion x', 'Y': 'Posicion y', 
                                'player': 'Jugador', 'situation': 'Situacion del disparo', 'season': 'Temporada', 
                                'shotType': 'Tipo Tiro', 'date': 'Fecha del partido', 'player_assisted': 'Jugador Asistente', 
                                'lastAction': 'Ultima accion'})

big5_renamed
```

```python
big5_renamed['Posicion x'] = big5_renamed['Posicion x'].str.replace('.', ',', regex=False)
big5_renamed['Posicion y'] = big5_renamed['Posicion y'].str.replace('.', ',', regex=False)
big5_renamed['xG'] = big5_renamed['xG'].str.replace('.', ',', regex=False)
big5_renamed
```

```python
# Reemplazar caracteres raros
from unidecode import unidecode
def limpiar_texto(texto):
    if isinstance(texto, str):
        texto = unidecode(texto)  # Eliminar acentos
        texto = texto.replace("&#039;", "")  # Eliminar apóstrofes
    return texto

big5_renamed['Jugador'] = big5_renamed['Jugador'].apply(limpiar_texto)
big5_renamed['Jugador Asistente'] = big5_renamed['Jugador Asistente'].apply(limpiar_texto)
```
